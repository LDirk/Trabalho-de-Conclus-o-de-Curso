```python
df = pd.read_csv("base/final_dataset.csv")
```

```python
target_counts = df["precipitation"].value_counts()

print("Frequência absoluta da variável precipitação:")
print(target_counts)
```

```python
target_prop = df["precipitation"].value_counts(normalize=True) * 100
print("\nFrequência relativa (%) da variável precipitação:")
print(target_prop.round(2))
```

```python
# Associações. 
```

```python
resultados = []
from scipy.stats.contingency import association

for col in df.columns:
    if col != "precipitation":
        tabela = pd.crosstab(df[col], df["precipitation"])

        res = chi2_contingency(tabela)
        v = association(tabela, method="cramer")

        resultados.append((col, v, res.pvalue))

assoc_df = pd.DataFrame(resultados, columns=["variavel", "cramers_v", "p_value"])
assoc_df = assoc_df.sort_values("cramers_v", ascending=False)

print(assoc_df)
```

```python

tabelas_percentuais = {}

for col in df.columns:
    if col != "precipitation":
        tabela_pct = pd.crosstab(df[col], df["precipitation"], normalize="index") * 100
        tabelas_percentuais[col] = tabela_pct.round(2)

for var, tabela in tabelas_percentuais.items():
    print(f"\n{'='*60}")
    print(f"Tabela percentual por linha: {var} x precipitation")
    print(tabela)
```
